```python
import psycopg2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# Connection setup
def get_connection():
    return psycopg2.connect(
        host="localhost",
        port=5433,
        database="clinicalmind",
        user="clinicalmind",
        password="clinicalmind123"
    )

def run_query(sql):
    conn = get_connection()
    df = pd.read_sql_query(sql, conn)
    conn.close()
    return df

print("✅ Connected to ClinicalMind PostgreSQL!")
```

```python
import os

# Create charts directory using absolute path
charts_dir = os.path.join(os.path.expanduser('~'), 'clinicalmind', 'sql', 'charts')
os.makedirs(charts_dir, exist_ok=True)
print(f"Charts directory: {charts_dir}")
```

```python
query1 = """
    SELECT 
        medical_specialty,
        COUNT(*) as total_notes,
        ROUND(AVG(word_count), 0) as avg_word_count,
        ROUND(AVG(char_count), 0) as avg_char_count,
        MAX(word_count) as max_word_count,
        MIN(word_count) as min_word_count
    FROM clean.clinical_notes
    GROUP BY medical_specialty
    ORDER BY total_notes DESC
    LIMIT 15
"""

df_specialty = run_query(query1)
display(df_specialty)

plt.figure(figsize=(12, 6))
sns.barplot(data=df_specialty, x='total_notes', y='medical_specialty', 
            hue='medical_specialty', legend=False, palette='Blues_r')
plt.title('Top 15 Medical Specialties by Number of Notes')
plt.xlabel('Total Notes')
plt.ylabel('Medical Specialty')
plt.tight_layout()
plt.savefig(os.path.join(charts_dir, 'specialty_distribution.png'), dpi=150)
plt.show()
print("✅ Query 1 complete!")
```

```python
# Query 2: CTE - Rank specialties by note volume and word count
query2 = """
    WITH specialty_metrics AS (
        SELECT 
            medical_specialty,
            COUNT(*) as total_notes,
            ROUND(AVG(word_count), 0) as avg_words,
            ROUND(AVG(char_count), 0) as avg_chars,
            SUM(word_count) as total_words
        FROM clean.clinical_notes
        GROUP BY medical_specialty
    ),
    ranked_specialties AS (
        SELECT *,
            RANK() OVER (ORDER BY total_notes DESC) as volume_rank,
            RANK() OVER (ORDER BY avg_words DESC) as complexity_rank
        FROM specialty_metrics
    )
    SELECT 
        medical_specialty,
        total_notes,
        avg_words,
        total_words,
        volume_rank,
        complexity_rank
    FROM ranked_specialties
    ORDER BY volume_rank
    LIMIT 15
"""

df_cte = run_query(query2)
display(df_cte)
print("✅ Query 2 (CTE + Window Functions) complete!")
```

```python
# Query 3: Window functions - running totals and percentiles
query3 = """
    WITH specialty_counts AS (
        SELECT 
            medical_specialty,
            COUNT(*) as total_notes,
            ROUND(AVG(word_count), 0) as avg_words
        FROM clean.clinical_notes
        GROUP BY medical_specialty
    )
    SELECT 
        medical_specialty,
        total_notes,
        avg_words,
        SUM(total_notes) OVER (ORDER BY total_notes DESC) as running_total,
        ROUND(100.0 * total_notes / SUM(total_notes) OVER (), 2) as pct_of_total,
        ROUND(100.0 * SUM(total_notes) OVER (ORDER BY total_notes DESC) / SUM(total_notes) OVER (), 2) as cumulative_pct,
        NTILE(4) OVER (ORDER BY avg_words DESC) as word_count_quartile
    FROM specialty_counts
    ORDER BY total_notes DESC
    LIMIT 15
"""

df_window = run_query(query3)
display(df_window)

# Visualize cumulative distribution
plt.figure(figsize=(12, 5))
plt.plot(df_window['medical_specialty'], df_window['cumulative_pct'], marker='o', color='steelblue')
plt.xticks(rotation=45, ha='right')
plt.title('Cumulative % of Notes by Specialty (Top 15)')
plt.ylabel('Cumulative %')
plt.xlabel('Medical Specialty')
plt.axhline(y=80, color='red', linestyle='--', label='80% threshold')
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(charts_dir, 'cumulative_distribution.png'), dpi=150)
plt.show()
print("✅ Query 3 (Running totals + Percentiles) complete!")
```

```python
# Query 4: Subquery - specialties with above average word count
query4 = """
    SELECT 
        medical_specialty,
        avg_word_count,
        total_notes,
        ROUND(avg_word_count - overall_avg, 0) as diff_from_avg,
        CASE 
            WHEN avg_word_count > overall_avg * 1.2 THEN 'Much Above Average'
            WHEN avg_word_count > overall_avg THEN 'Above Average'
            WHEN avg_word_count > overall_avg * 0.8 THEN 'Below Average'
            ELSE 'Much Below Average'
        END as complexity_category
    FROM (
        SELECT 
            medical_specialty,
            ROUND(AVG(word_count), 0) as avg_word_count,
            COUNT(*) as total_notes,
            ROUND(AVG(AVG(word_count)) OVER (), 0) as overall_avg
        FROM clean.clinical_notes
        GROUP BY medical_specialty
    ) subquery
    ORDER BY avg_word_count DESC
    LIMIT 15
"""

df_subquery = run_query(query4)
display(df_subquery)

# Visualize
colors = {
    'Much Above Average': '#2ecc71',
    'Above Average': '#3498db', 
    'Below Average': '#e67e22',
    'Much Below Average': '#e74c3c'
}
df_subquery['color'] = df_subquery['complexity_category'].map(colors)

plt.figure(figsize=(12, 6))
plt.barh(df_subquery['medical_specialty'], df_subquery['avg_word_count'], 
         color=df_subquery['color'])
plt.title('Average Word Count by Specialty (Top 15)')
plt.xlabel('Average Word Count')
plt.tight_layout()
plt.savefig(os.path.join(charts_dir, 'word_count_by_specialty.png'), dpi=150)
plt.show()
print("✅ Query 4 (Subqueries + CASE) complete!")
```

```python
# Query 5: JOIN - Compare raw vs clean layer
query5 = """
    SELECT 
        r.medical_specialty,
        COUNT(r.id) as raw_count,
        COUNT(c.id) as clean_count,
        COUNT(r.id) - COUNT(c.id) as dropped_records,
        ROUND(100.0 * COUNT(c.id) / COUNT(r.id), 2) as retention_rate
    FROM raw.clinical_notes r
    LEFT JOIN clean.clinical_notes c ON r.id = c.raw_id
    GROUP BY r.medical_specialty
    ORDER BY raw_count DESC
    LIMIT 15
"""

df_join = run_query(query5)
display(df_join)

plt.figure(figsize=(12, 6))
x = range(len(df_join))
plt.bar(x, df_join['raw_count'], alpha=0.7, label='Raw Count', color='steelblue')
plt.bar(x, df_join['clean_count'], alpha=0.7, label='Clean Count', color='green')
plt.xticks(x, df_join['medical_specialty'], rotation=45, ha='right')
plt.title('Raw vs Clean Record Count by Specialty')
plt.ylabel('Number of Records')
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(charts_dir, 'raw_vs_clean.png'), dpi=150)
plt.show()
print("✅ Query 5 (Complex JOIN) complete!")
```

```python
# Query 6: CASE - Categorize notes by length
query6 = """
    WITH note_categories AS (
        SELECT 
            medical_specialty,
            specialty_category,
            word_count,
            CASE 
                WHEN word_count < 100 THEN 'Very Short (<100 words)'
                WHEN word_count < 300 THEN 'Short (100-300 words)'
                WHEN word_count < 600 THEN 'Medium (300-600 words)'
                WHEN word_count < 1000 THEN 'Long (600-1000 words)'
                ELSE 'Very Long (>1000 words)'
            END as length_category
        FROM clean.clinical_notes
    )
    SELECT 
        length_category,
        COUNT(*) as total_notes,
        ROUND(100.0 * COUNT(*) / SUM(COUNT(*)) OVER (), 2) as percentage,
        ROUND(AVG(word_count), 0) as avg_words
    FROM note_categories
    GROUP BY length_category
    ORDER BY avg_words
"""

df_categories = run_query(query6)
display(df_categories)

plt.figure(figsize=(10, 6))
plt.pie(df_categories['total_notes'], 
        labels=df_categories['length_category'],
        autopct='%1.1f%%',
        colors=['#e74c3c', '#e67e22', '#3498db', '#2ecc71', '#9b59b6'])
plt.title('Distribution of Clinical Notes by Length')
plt.tight_layout()
plt.savefig(os.path.join(charts_dir, 'note_length_distribution.png'), dpi=150)
plt.show()
print("✅ Query 6 (CASE categorization) complete!")
```
